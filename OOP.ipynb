```python
"""
The self parameter is a reference to the current instance of the class.
It does not have to be named self, but it has to be the first parameter of any function in the class.
"""

class FinInst():
    def __init__(self, smbl, prc):
        self.symbol = smbl
        self.price = prc


a = FinInst('aapl', 600)

a.symbol, a.price
```

```python
class FinInst():
    def __init__(self, symbl, prc):
        self.symbol = symbl
        self.__price = prc  # private

    def get_price(self):
        return self.__price

b = FinInst('aapl', 10)

b.get_price() 
```

```python
b.__price   # is a private attribute, so this will result in an error
```

```python
b._FinInst__price  # If the class name is prepended with a single leading underscore, direct access and manipulation are still possible.
```

```python
b._FinInst__price = 500
b._FinInst__price
```

## Inheritance

```python
# Inheritance
class FinInstrument(FinInst):  # inherits FinInst
    def set_price(self, price):
        self._FinInst__price = price # !!!! price is attribute of the FinInst not FinInstrumet object, thats why it has to have the prefix
```

```python
inst = FinInstrument('aapl', '800')
inst.symbol
```

```python
inst.get_price() # function not defined for FinInstrument but only for FinInst
```

```python
inst.set_price(9000)
inst.get_price()
```

## Agregation

```python
class PortPos():
    def __init__(self, FinInstrument, size):
        self.size = size
        self.position = FinInstrument  # uses a different class

    def value_portfolio(self):
        return self.size * self.position.get_price() #can aggregate across different classes
```

```python
p = PortPos(inst, 10)
p.size
```

```python
p.position.get_price()
```

```python
p.position.symbol
```

```python
p.value_portfolio()
```
