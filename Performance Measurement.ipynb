## Performence Measurement

```python
import timeit
```

```python
timeit.timeit("4 + 4") #returns how many seconds it takes to execute the code 1mil times
```

```python
timeit.timeit("4 + 4", number = 1) #we can set up how many times it should run, if I wanna measure how long it takes to run it once, set it to 1
```

```python
timeit.timeit(
    """
a = 5
[i**i for i in range(10)]
    """
)
```

```python
timeit.timeit("4 + 4", number = 1) #we can set up how many times it should run, if I wanna measure how long it takes to run it once, set it to 1
```

```python
a = 5
b = 7
timeit.timeit("a + b", number = 1) #it can not access variables outside of the function, but is fixable with  globals = globals()
```

```python
a = 5
b = 7
timeit.timeit("a + b", number = 1, globals = globals()) #we can set up how many times it should run, if I wanna measure how long it takes to run it once, set it to 1
```
