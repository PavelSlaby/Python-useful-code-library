## Error Handling with try/except

### Best practice:
- Do NOT wrap everything in try/except.
- validate inputs/assumptions early

```python
# executed if any error

try:
    print(10 / 0)
except:  # executed if any error - but not a best practice as, it silences the errors - no matter what error happens, it will always go into the below print statement
    print("Yo, error message")
```

```python
# Best practice is to either catch only specific exemptions, or catch them all and log them all. 

try:
    print(10 / 0)
except Exception as e:  # executed if any error - but not a best practice as, it silences the errors - no matter what error happens, it will always go into the below print statement
    print(f"Yo, error message: {e}")
```

```python
# Else runs only if all went well, and finally runs in any scenario

try:
    1/1
except ZeroDivisionError:
    print('error')
else:
    print('printed if all good')
finally:
    print('printed in any scenario')
```

```python
try:
    1 / 0
except Exception as e:
    print(type(e).__name__) # prints only the class name
    print(e)                # prints the error message


```

```python
# we can raise specific exceptions

if 2 > 1:
    raise Exception("Inequality holds")
```

```python
# we can raise specific exceptions

if 2 > 1:
    raise ValueError("Value error was raised")
```
