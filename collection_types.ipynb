## Collection/Container Types

- <b>Sequence Types: </b> list, tuple, range
- <b>Mapping Type:</b> dict
- <b>Set Types:</b>  set, frozenset


-  List
   - ordered
   - changeable
   - allows duplicate members
-  Tuple
    - ordered
    - unchangeable (but lists and other types that are changeble can be changed if within the tuple).
    - allows duplicate members.
-  Set
      - unordered
      - unindexed
      - no duplicate members.
-  Dictionary
       - unordered
       - changeable
       - indexed.
       - no duplicate members. If here are duplicates, then it will always take the first one



## Tuples

```python
a = [5] # will be a list as it uses BRACKETS
e = (5,) # will be a tuple as it uses PARENTHESES

type(a), type(e)
```

```python
b = (5)  # will not be a tuple - because it is only one item
c = (5,) # will be a tuple, as it contains the comma

type(b), type(c)
```

```python
t = (a, 'aba', 12, 1.20)
t1 = a, 'aba', 12, 1.20  # no need for parentheses, but it is a best practice

type(t), type(t1)
```

```python
print( (*t1, 6, 5) ) # asterisks "unpacks" the tuple, which has the same effect as the following:

print( t1 + (6, 5) ) 

print(  (t1, 6, 5) ) # but this is not unpacked 
```

```python
print( t[1]          )
print( t[0]          ) 
print( t[0:1][0]     ) 
print( type(t[3])    ) 
print( t.index(12)   ) 
print( t.count(12)   ) 
```

```python
t[0] = 5 # expected error, tuples are unchangeable
```

## Lists

```python
list( (5,6) ) # convert anything to list
```

```python
tuple( [5,6] ) # convert anything to tuple
```

```python
l = [1, 2, 3, 'date']
print(l[2])
```

```python
l.append([2, 'ahoj']) # appends by the list
l
```

```python
l.extend([2, 'ahoj']) # extends the list
l
```

```python
l.insert(1, 'first position') # inserts at first position
l
```

```python
l.remove('ahoj')  # removes only the first ocurence
l
```

```python
l[3] = 5
l[2:4]
```

```python
l[1:5:2] = 'gu'  # assigns the sequence on the right to the list on the left, one sign at a time, as long as there are enough signs
l
```

```python
l.count('2')
```

```python
l.count(2)
```

```python
l.reverse() # permanent change
l
```
